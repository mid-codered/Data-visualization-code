```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'sales': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500],
    'profit': [200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100],
    'expenses': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]})

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(data['year'], data['sales'], label='Sales', color='blue')
plt.bar(data['year'], data['profit'], label='Profit', color='green', bottom=data['sales'])
plt.bar(data['year'], data['expenses'], label='Expenses', color='red', bottom=data['sales'] + data['profit'])

# Titles and labels
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales, Profit, and Expenses for Each Year')
plt.legend()

plt.show()
```
