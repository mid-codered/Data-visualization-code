```python
import matplotlib.pyplot as plt

# Data
outer_labels = ['A', 'B', 'C']
inner_labels = ['D', 'E', 'F', 'G']
outer_sizes = [30, 40, 30]
inner_sizes = [[10, 10, 5, 5], [20, 10, 5, 5], [5, 20, 5, 0]]

# Plot
fig, ax = plt.subplots()

# Outer pie chart
outer_colors = ['lightcoral', 'lightskyblue', 'lightgreen']
ax.pie(outer_sizes, labels=outer_labels, colors=outer_colors, startangle=100)

# Inner pie chart
inner_colors = ['red', 'blue', 'green', 'gray']
for i, size in enumerate(outer_sizes):
    inner_radius = size * 0.9
    ax.pie(inner_sizes[i], radius=inner_radius, labels=inner_labels, colors=inner_colors, startangle=100)


# Equal aspect ratio ensures the pie charts are circular
ax.axis('equal')

# Show the plot
plt.show()
```
